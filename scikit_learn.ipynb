## Pre-Processing


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load Data
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler() # subtracts mean and divides by variance -> normal distribution

# Scaler must be trained on the data before it transforms
X_train_scaled = scaler.fit_transform(X_train)    # Fit -> Trains the scaler on the data, Transform -> does the scaling
X_test_scaled = scaler.transform(X_test)    # Does not need to be re-fit on the same data -> Just Transform
```

```python
# Manual scaling with NumPy
import numpy as np

(X_train - np.mean(X_train, axis=0)) / np.std(X_train, axis=0)
```

```python
# Min / Max Scaling -> Normalizes data
from sklearn.preprocessing import MinMaxScaler    # NO NEGATIVE NUMBER, only on range [0,1]

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)    # Fit -> Trains the scaler on the data, Transform -> does the scaling
X_test_scaled = scaler.transform(X_test)    # Does not need to be re-fit on the same data -> Just Transform

X_train_scaled
```

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Many different datatypes, not all numerical
data = fetch_openml('car', as_frame=True).frame
data
```

```python
# Encode lug_boot & safety with OrdinalEncoder -> for Ordinal Attributes

columns_to_encode = ['lug_boot', 'safety']
encoder = OrdinalEncoder(categories=[ ## specifies the order for the columns
    ['small', 'med', 'big'],
    ['low', 'med', 'high']
])

data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

```python
data
```

```python
# Possible to reverse the encoding via the encoder.inverse_transform()
encoder.inverse_transform(data[columns_to_encode])
```

```python
# One Hot Encoder - Nominal Attributes
data = fetch_openml('adult', as_frame=True).frame
data
```

```python
from sklearn.preprocessing import OneHotEncoder
# Use to split Nominal attributes into binary attributes like -- | Occupation | -> | Cleaner | Manager | Clerical | ....

data.occupation.value_counts()
```

```python
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)   # what if we come accross an unknown data later?? 'ignore' it

encoded_values = encoder.fit_transform(data[['occupation', 'race']])    # encodes the attributes as 1 / 0
new_cols = encoder.get_feature_names_out(['occupation', 'race'])    # returns the attribute names as new column names
```

```python
encoded_values
```

```python
new_cols
```

```python
# create a new Dataframe with new values and columns as new cols, specifying the index as the same as previous dataset
df_encoded = pd.DataFrame(encoded_values, columns=new_cols, index=data.index)

# Merge the two dataframes, while dropping the 2 old categories we just encoded
data_final = pd.concat(
    [data.drop(columns=['occupation', 'race']), df_encoded], axis=1
)

data_final
```

## PCA - Principal Component Analysis
* high-dim dataset and reduce dimensionality
* find directions in n-dim space that explain the most variance and reduce dataset down to those dimensions

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# 28x28 images of hand drawn numbers
X, y = fetch_openml('mnist_784', return_X_y=True)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
pca = PCA(n_components=20)

X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
X_train.shape
```

```python
X_train_reduced.shape
```

```python
clf = LogisticRegression(max_iter=100)
clf.fit(X_train, y_train)

print(clf.score(X_test, y_test))
```

```python
clf = LogisticRegression(max_iter=100)
clf.fit(X_train_reduced, y_train)

print(clf.score(X_test_reduced, y_test))
```

```python
pca.explained_variance_ratio_
```

## Metrics

```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load the data
data = load_breast_cancer(as_frame=True).frame
X, y = load_breast_cancer(return_X_y=True)
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# instantiate the Scaler
scaler = StandardScaler()

# Scale the data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train_scaled, y_train)
```

```python
clf.score(X_test_scaled, y_test)    # Accuracy
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Create an array of predicitions to pass to accuracy score funtion
y_pred = clf.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

## Hyperparameter Tuning


```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Sets up 18 different configuration of hyperparameters to try
param_grid = {
    'n_estimators': [50, 100, 200],
    'min_samples_split':[2, 5],
    'max_depth': [None, 5, 10],
}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Instatiate the classifier
clf = RandomForestClassifier(n_jobs=-1) # use all CPU cores

# Create the Grid Cross validator and pass the custom grid and testing each of 18 configurations 3 times
grid = GridSearchCV(clf, param_grid, cv=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
best_clf = grid.best_estimator_
best_clf
```

```python
best_clf.score(X_test, y_test)
```
